### Converting Labels to numbers

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
cc[cc.select_dtypes(include='object').columns] = cc[cc.select_dtypes(include='object').columns].apply(le.fit_transform)
```

```python
cc.head()
```

```python
cc.info()
```

### Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X = cc.drop('Churn',axis=1)
y = cc['Churn']
```

```python
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_test.head()
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
type(y_train)
```

```python
pred_test = logreg.predict(X_test)
```

```python
pred_test
```

```python
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, pred_test)
```

```python
print(classification_report(y_test, pred_test))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
pred_dt = dt.predict(X_test)
```

```python
confusion_matrix(pred_dt, y_test)
```

```python
print(classification_report(pred_dt, y_test))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```
